```python
import pandas
import numpy as np
```

```python
def read_data(typeData, labelsInfo, imageSize, path):
    #Intialize x matrix
    x = np.zeros(shape=(len(labelsInfoTrain), imageSize))

    for (index, idImage) in enumerate(labelsInfo["ID"]): 
        #Read image file 
        #nameFile = path+"/"+typeData+"Resized/"+idImage+".Bmp"
        nameFile="%s/%sResized/%s.Bmp" %(path,typeData,idImage)
        img = Image.open(nameFile).convert('L')

        #Transform image matrix to a vector and store 
        #it in data matrix 
        x[index, :] = np.reshape(img, (1, imageSize))
    
     
    return x
```
